```python
import pandas as pd
import dash
from dash import html
import pandas as pd
from dash import dcc
from dash.dependencies import Input, Output, State
import plotly.express as px
import base64
import numpy as np
```

```python
df = pd.read_csv('tobacco.csv')
for col in df.columns:
    if col not in ['Year', 'State', 'Location 1']:
        df[col] = df[col].str.rstrip('%').astype('float')/100.0

        
df = df[(df['State'] != 'Nationwide (States, DC, and Territories)') & (df['State'] != 'Nationwide (States and DC)')]
df['all_tobacco'] = df['Smoke some days'] + df['Smoke everyday']
df['state_code'] = df.apply(lambda x: us_state_to_abbrev[x['State']], axis=1)
df_grouped = df.groupby('Year')['Smoke some days', 'Smoke everyday', 'Former smoker', 'Never smoked'].agg('sum')


fig =  px.choropleth(data_frame=df, color_continuous_scale=px.colors.sequential.Blues, 
                    hover_data=['State'],	color='all_tobacco', 
                     locationmode='USA-states', locations='state_code', 
                     scope='usa', template='plotly_dark')
fig_bar = px.bar(
        data_frame=df,
        x='State',
        y=['Smoke some days', 'Smoke everyday', 'Former smoker', 'Never smoked'],
        barmode='group',
        template='plotly_dark'
)

fig_line = px.line(
        data_frame=df_grouped,
        x=df_grouped.index,
        y=['Smoke some days', 'Smoke everyday', 'Former smoker', 'Never smoked'],
        template='plotly_dark'
)

fig_line.update_traces(mode="markers+lines", hovertemplate=None)
fig_line.update_layout(hovermode="x")

app = dash.Dash(__name__)

@app.callback(
    dash.dependencies.Output('geo_graph', 'figure'),
    dash.dependencies.Output('bar_graph', 'figure'),
    dash.dependencies.Input('my-slider', 'value')
    )
def update_graph(year_value):
    fig1 = px.choropleth(data_frame=df[df['Year']==year_value],
                    hover_data=['State'],
                    color_continuous_scale=px.colors.sequential.Blues, color='all_tobacco', 
                    locationmode='USA-states', locations='state_code', scope='usa', template='plotly_dark'
                    )

    fig2 = px.bar(
        data_frame=df[df['Year'] == year_value],
        x='State',
        y=['Smoke some days', 'Smoke everyday', 'Former smoker', 'Never smoked'],    
        template='plotly_dark',
        )
    return fig1, fig2

image_filename = 'download.jpg'
encoded_image = base64.b64encode(open(image_filename, 'rb').read())


colors = {
    'background': 'black'
}

app.layout = html.Div(style={'backgroundColor': colors['background']}, children=[
    html.Div([
        
    html.Img(src='data:image/png;base64,{}'.format(encoded_image.decode()),style={'width':'400px','margin': 'auto'})
,
        html.H1("Tobacco Used in US between 1995-2010:",style={'textAlign':'center','margin': 'auto','color':'white',"text-decoration": 'underline'})      
             ]),
    html.Br(),
    html.Br(),
    html.Br(),
    html.Div([
        
        dcc.Slider(
            id='my-slider',
            min = df['Year'].min(),
            max = df['Year'].max(),
            value = df['Year'].min(),
            marks = {str(year):str(year) for year in df['Year'].unique() }
            )
        
        ]),
    dcc.Graph(id='bar_graph', figure=fig_bar),
    
    html.H1("A map Shows the use of tobacco in Each State in US :",style={'textAlign':'center','margin': 'auto','color':'white',"text-decoration": 'underline'}),
        html.Br(),
    html.Br(),
    dcc.Graph(id='geo_graph', figure=fig),
    
    html.H1("Types of Smokers :",style={'textAlign':'center','margin': 'auto','color':'white',"text-decoration": 'underline'}),
        html.Br(),
    html.Br(),
    
    dcc.Graph(id='line_graph', figure=fig_line)
    
    ])

if __name__ == '__main__':
    app.run_server()
```
